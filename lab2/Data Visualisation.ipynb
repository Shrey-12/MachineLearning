```python
import random
import numpy as np
np.random.seed(100)
X = np.random.uniform(1, 3, 6)
Y = 1.7 + 0.5 * X
```

```python
Y
```

```python
alpha = 0.1
iterations=3
```

```python
m=0.5
c=1.7
```

```python
import matplotlib.pyplot as plt
fig,axes=plt.subplots(nrows=iterations,ncols=1,figsize=(6,10))
def compute_cost(m, c):
    predictions = m * X + c
    cost = np.sum((predictions - Y)**2) / (len(Y))
    return cost

# Gradient Descent
for i,ax in enumerate(axes):
    m_gradient = np.sum(2*(m * X + c - Y) * X) / len(Y)
    c_gradient = np.sum(2*(m * X + c - Y)) / len(Y)
    m -= alpha * m_gradient
    c -= alpha * c_gradient
    cost = compute_cost(m, c)
    
    ax.scatter(X,Y,label='data points')
    ax.plot(X,m*X+c,color='orange',label='Regression Line')
    ax.set_ylabel('Y')
    ax.set_title(f"Iteration {i+1}: Cost = {cost:.4f}")
    ax.legend()

plt.show()
print("Final values of m and c:", m, c)
```
